# What is Data Visualization ?
Data visualization is a technique that uses an array of static and interactive visuals within a specific context to help people understand and make sense of large amounts of data. The data is often displayed in a story format that visualizes patterns, trends and correlations that may otherwise go unnoticed

## Data visualization in Machine Learning
Data visualization is an important skill in applied statistics and machine learning.
Statistics does indeed focus on quantitative descriptions and estimations of data. Data visualization provides an important suite of tools for gaining a qualitative understanding

```python
import pandas as pd
```

```python
df= pd.read_csv("heart_disease_dataset.csv")
```

```python
df
```

```python
df.info()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
t=df["age"]
s=df["thalach"]
```

```python
t[0]
```

```python
fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='thalach(s)', ylabel='Age',
       title='Simple Plot')
ax.grid()

# fig.savefig("test.png")
plt.show()
```

```python
plt.plot(t,s)

plt.title('Heart Disease')
plt.ylabel("Age")
plt.xlabel('bp')

plt.show()
```

#### Scatter Plot

```python
plt.scatter(t,s)
plt.title('Graph')
plt.ylabel('Age')
plt.xlabel('bp')

plt.show()
```

#### 2D histogram

```python
plt.hist2d(t,s)
plt.title('Heart Disease')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
z=df["chol"]
```

#### Fill plot

```python

plt.fill(t, s)
plt.title("Fill plot")
plt.show()
```

#### Bar plot

```python
fig, ax = plt.subplots()
ax.barh(s, z)
```

### Pie Chart

```python
t[0]
```

```python
s[0]
```

```python
z[0]
```

```python



labels = 'Machine Learning', 'Deep Learning ', 'Data Science'
sizes = [t[0], s[0], z[0]]
explode = (0, 0.1, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=False, startangle=None)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
